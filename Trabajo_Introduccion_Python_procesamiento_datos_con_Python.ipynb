```python
# 1. Calculo el Máximo Común divisor de dos números
# Utilizo el coding: utf-8
 
# Defino la función de cálculo
def mcd(a, b):
    resto = 0
    while(b > 0):
        resto = b
        b = a % b
        a = resto
    return a
 
# Solicito los dos números que vamos a calcular el Máximo Cómun Divisor
num1 = int(input("Introduce el primer numero: "))
num2 = int(input("Introduce el segundo numero: "))
 
print("El máximo común divisor de ", num1," y ", num2," es ", mcd(num1, num2))
```

```python
# 2. Calculo el Mínimo Común múltipo de dos números
# Utilizo el coding: utf-8
 
# Defino la función de cálculo
def lcm(x, y):
# La función recibe dos enteros y devuelve el multiplo comum
 
# Cogo el mayor de los dos
   if x > y:
       mayor = x
   else:
       mayor = y
 
   while(True):
       if((mayor % x == 0) and (mayor % y == 0)):
           lcm = mayor
           break
       mayor += 1
 
   return lcm
 
# Solicito los dos numeros
num1 = int(input("Introduce el primer numero: "))
num2 = int(input("Introduce el segundo numero: "))
 
print("El minimo comun multiplo de ", num1," y ", num2," es ", lcm(num1, num2))
```

```python
#1 Leer el fichero proporcionado.

# archivo-entrada.py
fich_entrada = open("FicheroALeer.txt","r")
fich_entrada.close()

#2 Contar el número de líneas del fichero.

fname = "FicheroALeer.txt"
count = 0
with open(fname, 'r', encoding="utf-8", errors="ignore") as f:
    for line in f:
        count += 1
print("El número total de lineas del Fichero es:", count)

#3 Contar el número de veces que aparece el artículo «el» en el fichero.
    #Nota: elemento contiene los caracteres «el» pero no debe contar puesto que no contiene a el artículo «el».
     
fname = "FicheroALeer.txt"
count2 = 0
with open(fname, "r", encoding="utf-8", errors="ignore") as f:
    for line in f:
        cadena = line.split()
        for elemento in cadena:
            if elemento == "el":
                count2 += 1
print("El número de veces que aparece el artículo el:", count2)
    
#4 Eliminar las tildes del fichero y convertir todas las letras a minúsculas.

fname = "resultado.txt"
resultado = open(fname, "w")
resultado.write(str(count)+"\n")
resultado.write(str(count2)+"\n")
replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
    )

fname = "FicheroALeer.txt"
with open(fname, "r", encoding="utf-8", errors="ignore") as f: 
    for line in f:
        for a, b in replacements:
            line = line.lower().replace(a, b)
        print(line)
        resultado.write(line)

#5 Crear un fichero de texto con el nombre «resultado.txt» 
    #que contendrá en la primera línea el número de líneas del fichero,
    #en la segunda línea el número de veces que aparece el artículo «el»,
    # y a partir de la tercera línea el texto original todo en minúsculas y sin tilde.
    
resultado.close() 
```

```python
#1 Cargar en el entorno el «poblacionMunicipios.csv» respetando las tildes y caracteres españoles.
import pandas as pd
import os
#df = pd.read_csv('poblacionMunicipios.csv', sep=',')
df = pd.read_csv('poblacionMunicipios.csv',converters={'Poblacion': int})
#2 Eliminar las filas correspondientes a municipios sin población.
df = df[df['Poblacion'] > 0]
#print(df)
#3 Imprimir el total de la población de todos los municipios.
print(df['Poblacion'].sum())
#4 Crear un DataFrame de Pandas donde
    #la primera columna sean todas las provincias que aparecen en el archivo,
    #la segunda columna el número total de habitantes por provincia, 
    #la tercera la desviación típica en el número total de habitantes por provincia y
    #la cuarta el número total de municipios por provincia.
#print(df.groupby(['CodProvincia','Provincia'])['Poblacion'].sum())
#print(df.groupby(['CodProvincia','Provincia'])['Poblacion'].std())
#print(df.groupby(['CodProvincia','Provincia'])['Poblacion'].count())
df_poblacion_provincia=df.groupby(['Provincia'])['Poblacion'].agg(['sum','std','count']).rename(columns={'sum':'Poblacion_Total','std':'Desviacion_Tipica','count':'Numero_Municipios'})
#print(df_poblacion_provincia)
#5 Cargar el archivo «CP_Municipios.csv»
df_cp = pd.read_csv('CP_Municipios.csv',converters={'CP': str})
#6 Empleando el código de provincia y el código de municipio, cruza los dos datasets comentados hasta el momento.
    #el resultado final debe ser un DataFrame.
df2=pd.merge(df, df_cp, on=['CodProvincia','CodMunicipio'])    
#print(df2)
#7 Generar un CSV llamado «faltan.csv» con todos los códigos postales que estén
    #en «CP_Municipios.csv» pero que no tengan en «poblacionMunicipios.csv»
    #ningún municipio con el que poder cruzar. Si todo cruza correctamente el
    #archivo «faltan.csv» estará vacío. Se puede decir que «faltan.csv» son los
    #códigos postales que se han perdido en el cruce a la hora de considerar la población.
df_merge_cp=pd.merge(df_cp,df, how='outer', indicator='union')
df_faltan=df_merge_cp[df_merge_cp.union=='left_only']
#print(df_faltan)
export_csv = df_faltan.to_csv("faltan.csv",encoding='utf-8', index=False,columns=['CP'])
#8 Agrupar la información por código postal. Se quiere un DataFrame con las siguientes columnas:
    #Código postal.
    #Número de municipios que tienen dicho código postal asignado.
    #Población: se calculará como la suma de la población de todos los municipios que incluyen a dicho código postal.
    #Provincia: a la que está asignada el código postal.
df2.groupby(['CP'])['Poblacion'].agg(['count','sum']).rename(columns={'sum':'Poblacion_Municipios','count':'Numero_Municipios'})
```
